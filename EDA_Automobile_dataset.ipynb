```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# to load the data using pd.read_csv('the path')

df =  pd.read_csv('/content/Automobile_data.csv')
df.head(10)
```

```python
# since there are missing column names , we are adding column names
colmn=['symboling','normalized_losses','make','fuel_type','aspiration','no_of_doors','body_type','drive_wheels_engine','location','wheel_base','length','width','height','curb_weight','engine_type','num_of_cylinders','engine_size','fuel_system','bore','stroke','compression_ratio','horsepower','peak_rpm','city_mpg','highway_mpg','price']
df.columns = colmn
df.head()
```

```python
# now to check for missing values
df.isnull().sum()
```

```python
df.info()
```

```python
for colmn in df.columns:
    df[colmn].replace({'?':np.nan},inplace=True)\

df.head()
```

```python
df.isnull().sum()
```

```python
# now going to visualise the missing values
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
```

```python
num_colmn = ['normalized_losses','bore','stroke','horsepower','peak_rpm','price']
for colmn in num_colmn:
  df[colmn]=pd.to_numeric(df[colmn])
  df[colmn].fillna(df[colmn].mean(), inplace=True)

df.head()
```

```python
# making a heat map
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),cbar=True,annot=True,cmap='Blues')
```

```python
# how does the horse power realated with the price
plt.figure(figsize=(10,10))
plt.scatter(x='horsepower',y='price',data=df)
plt.xlabel('horsepower')
plt.ylabel('price')
plt.legend()
```

from the above plot we observe that the outliers are between 200 to 300

```python
# to know the relation between engine size & price
plt.figure(figsize=(10,10))
plt.scatter(x='engine_size',y='price',data=df)
plt.xlabel('Engine size')
plt.ylabel('Price')
```

from the above plot we observe that the outliers lie between 200 to 300

```python
# here we see how the highway_mpg affects the price
plt.figure(figsize=(10,10))
plt.scatter(x='highway_mpg',y='price',data=df)
plt.xlabel('Higway mpg')
plt.ylabel('Price')
```

from the above code we can see that with increase in highway mpg there is a decrease in price



```python
# relation between no.of doors and price
value_of_no_of_doors = df.no_of_doors.value_counts()
print(value_of_no_of_doors)
sns.boxplot(x='price',y='no_of_doors',data=df)
```
